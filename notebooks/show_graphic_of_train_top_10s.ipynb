```python
import pandas as pd
import os

current_dir = os.getcwd()
relative_path = '../models/top10/'

# Define the directory where your CSV files are located
directory = os.path.join(current_dir, relative_path)

# Initialize a list to store the DataFrames
dfs = []
counter = 0

# Loop through all the CSV files in the directory
for filename in os.listdir(directory):
    if filename.endswith('.csv'):
        
        # print(filename)
        
        # Construct the full file path
        file_path = os.path.join(directory, filename)
        
        # Load the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Add a new column indicating the source CSV file
        df['source'] = os.path.splitext(filename)[0]
        # df['source'] = ids[counter]
        
        counter += 1
        
        # Append the DataFrame to the list
        dfs.append(df)

# Concatenate all DataFrames in the list into a single DataFrame
merged_df = pd.concat(dfs, ignore_index=True)

merged_df
```

```python
merged_df['source'] = merged_df['source'].astype('int64')
```

```python
import pandas as pd

# Assuming 'df' is your DataFrame and 'source' is the column you want to change
numbers = [65, 62, 68, 50, 56, 89, 74, 86, 91, 95, 1, 2, 11, 23, 7, 25, 40, 31, 46, 26, 113, 104, 116, 119, 98, 115, 110, 109, 134, 137]
mapping = {i+1: num for i, num in enumerate(numbers)}

merged_df['source'] = merged_df['source'].map(mapping)
```

```python
final_merge = pd.merge(merged_df, config_df, left_on='source', right_on='Id')
```

```python
final_merge = final_merge.fillna("None")
final_merge = final_merge.drop('Id', axis=1)

final_merge['Learning Rate'] = final_merge['Learning Rate'].astype('string')
```
