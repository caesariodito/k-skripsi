```python
import pandas as pd
import matplotlib.pyplot as plt
import dtale
```

```python
# Create a function to plot unique value counts
def plot_unique_value_counts(df):
    unique_counts = []
    column_names = []

    for column in df.columns:
        unique_count = df[column].nunique()
        unique_counts.append(unique_count)
        column_names.append(column)

    plt.figure(figsize=(10, 6))
    plt.bar(column_names, unique_counts)
    plt.ylabel("Count")
    plt.title("Unique Value Counts in Columns")
    plt.show()
```

```python
file_path = "./csv_results/scenarios.csv"

df = pd.read_csv(file_path)
df = df.fillna("None")
df
```

```python
df["Learning Rate"] = df["Learning Rate"].astype(str)
```

```python
df.info()
```

```python
plot_unique_value_counts(df)
```

```python
file_path = "./csv_results/scenarios.csv"

df = pd.read_csv(file_path)
df = df.fillna("None")

df["Learning Rate"] = df["Learning Rate"].astype(str)

# PREPROCESS THE DATAFRAME TO SEE THE CORRELATION MATRIX FOR EACH CONFIGS

columns_to_take = ["Model", "Augmentasi", "Freeze", "Learning Rate", "Layer", "Optimizer", "Test", "Time (min)"]
columns_to_encode = ["Model", "Augmentasi", "Freeze", "Learning Rate", "Layer", "Optimizer"]
# columns_to_encode = ["Model", "Freeze"]

df_preprocessed = df[columns_to_take]

one_hot = pd.get_dummies(df_preprocessed[columns_to_encode])
df_preprocessed = df_preprocessed.drop(columns_to_encode, axis=1)
df_preprocessed = df_preprocessed.join(one_hot)

df_temp1 = df_preprocessed.pop("Test")
df_temp2 = df_preprocessed.pop("Time (min)")
df_preprocessed = df_preprocessed.join(df_temp1)
df_preprocessed = df_preprocessed.join(df_temp2)

df_preprocessed
```

```python
from dython.nominal import associations

associations(df_preprocessed, nom_nom_assoc="theil", figsize=(15, 15))
```
