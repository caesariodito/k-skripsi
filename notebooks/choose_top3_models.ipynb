```python
import pandas as pd

df = pd.read_csv('./top_10.csv')
df = df.fillna("None")
df
```

```python
# Find top models

df_sorted = df.sort_values(by=['Model', 'Test', 'Time (min)', 'Validation'], ascending=[True,False,True,False])

# Get top 3 rows for each model
top_models = df_sorted.groupby('Model').head(3)
```

```python
top_models
```

```python
# based on my Sheet "Metopen Thingy"
ids = [6, 1, 3, 17, 16, 12, 29, 21, 22]

# Create a new DataFrame with ids
df_ids = pd.DataFrame(ids, columns=['id'])

# Add a new column 'id' to top_models DataFrame
top_models = pd.concat([top_models.reset_index(drop=True), df_ids], axis=1)
```

```python
top_models.to_csv('chosen_models.csv')
```
